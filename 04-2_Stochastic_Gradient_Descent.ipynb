앞서 7개의 생선을 분류해내는 모델을 훈련하였다.  
이제는 훈련 데이터가 조금씩 새로 전달되고 새로 추가되는 생선의 종류도 존재한다. 어떻게 모델을 지속적으로 운영이 가능할까?  
기존의 훈련 데이터에 새로운 데이터를 계속 추가하여 매일매일 훈련하는건 어떨까?  
하지만 이는 시간이 지날수록 데이터가 늘어나고 서버의 추가 등 현실적인 제한점이 존재한다.  
또 다른 방법은 새로운 데이터를 추가할 때 이전 데이터를 버림으로서 훈련 데이터의 크기를 일정하게 유지하는 것이다.  
하지만 이 방법은 중요한 데이터의 손실이 발생할 수 있는 문제가 있다.  
위에서 언급한 두 방법은 이전에 훈련한 모델을 버리고 다시 새로운 모델을 훈련하는 방법이다. 앞서 훈련한 모델을 버리지 않고 새로운 데이터에 대해서만 조금씩 더 훈련할 수 없을까? 이렇게 할 수 있다면 훈련에 사용한 데이터를 모두 유지할 필요도 없고 앞서 학습한 생선을 까먹을 일도 없을 것이다.  
  
이런 식의 훈련 방식을 **점진적 학습**또는 온라인 학습이라고 부른다. 대표적인 점진적 학습 알고리즘은 **확률적 경사 하강법(Stochastic Gradient Decendent)**이다. 사이킷 런에서는 확률적 경사 하강법을 위한 클래스를 제공한다.

### 확률적 경사 하강법  
확률적 - 무작위하게, 랜덤하게  
경사 - 기울기  
하강법 - 내려가는 방법  
즉 경사를 따라 내려가는 방법을 의미한다.  

가장 가파른 경사를 따라 원하는 지점에 도달하는 것이 목표이며 걸음이 너무 크다면 경사를 따라 내려가지 못하고 오히려 올라갈 수도 있다. 그렇기 때문에 조금씩 내려오는 것이 중요하며 이렇게 내려오는 과정이 경사 하강법 모델을 훈련하는 것이다.  
확률적이라는 단어는 무엇을 뜻하는가?  
훈련 세트를 사용해 모델을 훈련하기 때문에 경사 하강법도 훈련 세트를 사용하여 가장 가파른 길을 찾을 것이다. 이때 전체 샘플을 사용하지 않고 하나의 샘플을 훈련 세트에서 랜덤하게 골라 가장 가파른 길을 찾는다. 훈련 세트에서 랜덤하게 하나의 샘플을 고르는 것이 확률적 경사 하강법이다.  
확률적 경사 하강법은 훈련 세트에서 랜덤하게 하나의 샘플을 선택하여 가파른 경사를 조금 내려간다. 그다음 훈련 세트에서 랜덤하게 또 다른 샘플을 하나 선택하여 경사를 조금 내려간다. 이런 방법으로 전체 샘플을 모두 사용할 때까지 계속한다.  
모든 샘플을 다 사용하고도 다 내려오지 못한다면 다시 처음부터 시작한다. 훈련 세트에 모든 샘플을 다시 채워넣고 다시 랜덤하게 하나의 샘플을 선택해 이어서 경사를 내려간다. 만족할만한 위치에 도달할 때까지 계속 내려간다.  
확률적 경사 하강법에서 훈련 세트를 한번 모두 사용하는 과정을 **에포크(epoch)**라고 부른다. 일반적으로 경사 하강법은 수십, 수백번 이상 에포크를 수행한다.

1개씩 말고 무작위로 몇 개의 샘플을 선택해서 경사를 따라 내려가면 더 안정적일 것이다. 이렇게 여러 개의 샘플을 사용해 경사 하강법을 수행하는 방식을 **미니배치 경사 하강법(minibatch gradient descent)라고 한다.  
한 번 경사로를 따라 이동하기 위해 전체 샘플을 사용할 수도 있다. 이를 **배치 경사 하강법(batch gradient descent)**이라고 한다. 전체 데이터를 이용하기 때문에 가장 안정적인 방법이지만 전체 데이터를 사용하면 그 만큼 컴퓨터 자원을 많이 사용하게 된다.

확률적 경사 하강법은 훈련 세트를 사용해 아래에 있는 최적의 장소로 조금씩 이동하는 알고리즘이므로 훈련 데이터가 모두 준비되어 있지 않고 매일매일 업데이트 되어도 학습을 계속 이어나갈 수 있다.

확률적 경사 하강법을 사용하는 알고리즘으로서 신경망 모델이 있다. 신경망 모델은 확률적 경사 하강법이나 미니배치 경사 하강법을 사용한다.

어디서 내려가야 하는가? 가장 빠른 길을 찾아 내려가려고 하는 이 산은 무엇일까? 이 산이 바로 손실 함수라고 부르는 것이다.

### 손실 함수  
**손실 함수(loss function)**는 어떤 문제에서 머신러닝 알고리즘이 얼마나 엉터리인지를 측정하는 기준이다.
